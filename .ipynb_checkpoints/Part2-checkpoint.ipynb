# 1) Study the job metadata. Extract the relevant information to describe the job’s attributes.

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt; plt.rcdefaults()
import plotly.express as px
```

```python
data = pd.read_csv('C:/Users/Minh/Desktop/Assignment/data.csv')
```

### What is the sector, sub-sector of each job?

```python
data[['Classification','SubClassification']].head(5)
     
```

```python
table1 = pd.pivot_table(data, columns=['SubClassification'] ,index=['Classification'], aggfunc={'SubClassification':'count'})
table1
```

### Where is the location of the job?

```python
data[["Location","Classification"]].head(5)
```

```python
table = pd.pivot_table(data, columns=['Classification'] ,index=['Location'], aggfunc={'Classification':'count'})
table
```

### Which is the range of salaries for each job?

```python
g4 = data[['HighestSalary','LowestSalary']].groupby(data['Classification']).agg({'HighestSalary' : ['max','min'],
                                                                            'LowestSalary' : ['max','min']})
g4
```

# 2) Study the market by locations. 

### What is the market size in each city? 

```python
a = data['HighestSalary'].groupby(data['Location']).mean()
a
```

### Which are the hottest job sectors in each city?

```python
b = pd.pivot_table(data, columns=['Classification'] ,index=['Location'], aggfunc={'Classification':'count'})
b
```

### Which range of salary is common in each city? Where are the employees more well-paid? 

```python
salary = data[['LowestSalary','HighestSalary']].groupby(data['Location']).mean().reset_index()
salary
```

### Can you detect the pattern of posting?

```python
data['Date'] = data['Date'].astype(str)
```

```python
data['Date'] = pd.to_datetime(data['Date'], errors='coerce', format='%Y-%m-%d')
```

```python
data['Date'] = data['Date'].apply(lambda x: x.strftime('%Y-%m-%d'))
```

```python
data.head(5)
```

```python
data['YearMonth'] = data['Date'].apply(lambda x : x.strftime('%Y-%m')) 
```

```python
data['YearMonth'] = pd.to_datetime(data['YearMonth'])
```

```python
data.head(5)
```

```python
table = pd.pivot_table(data, columns=['YearMonth'] ,index=['Classification'], aggfunc={'YearMonth':'count'})
table
```

```python
d = data['Classification'].groupby(data['YearMonth']).count().reset_index()
```

```python
d
```

# 3) Study the market by sectors.

### Which sectors keep the highest market share?

```python
data['Classification'].value_counts()
```

```python
data['RangeSalary'] = data['HighestSalary'] - data['LowestSalary']
data['RangeSalary']
```

```python
market = data['RangeSalary'].groupby(data['Classification']).mean().reset_index().sort_values('RangeSalary', ascending = False)
market
```

### In each sector, which sub-sectors are the main spotlights?

```python
main = pd.pivot_table(data, columns=['SubClassification'] ,index=['Classification'], aggfunc={'SubClassification':'count'},fill_value = 0)
main
```

### What is the salary range for each sector/sub-sector? Can you compare salary range between sectors/subsectors?

```python
# Range salary for each classification
salaryrange = data['RangeSalary'].groupby(data['Classification']).mean().reset_index().sort_values('RangeSalary', ascending = False)
salaryrange
```

```python
# Range salary for each classification
salaryrange1 = data['RangeSalary'].groupby(data['SubClassification']).mean().reset_index().sort_values('RangeSalary', ascending = False)
salaryrange1
```
